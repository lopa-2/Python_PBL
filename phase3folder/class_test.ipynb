```python
import joblib

model = joblib.load("rf_model.pkl")
sfs = joblib.load("sfs_selector.pkl")
features = joblib.load("feature_columns.pkl")
print(features)
```

```python
model = joblib.load("rf_model.pkl")
print(model)
```

```python
import pandas as pd

region = input("Enter region (e.g., USA): ")

Year = int(input("Enter Year: "))
athlete_count = int(input("Enter athlete count: "))
age_median = float(input("Enter age median: "))
height_median = float(input("Enter height median: "))
weight_median = float(input("Enter weight median: "))

Gold_prev = int(input("Enter previous Gold medals: "))
Silver_prev = int(input("Enter previous Silver medals: "))
Bronze_prev = int(input("Enter previous Bronze medals: "))

Total_prev = Gold_prev + Silver_prev + Bronze_prev

is_zero_prev = int(Total_prev == 0)

rolling_medals = float(input("Enter rolling medals: "))

gold_ratio = Gold_prev / (Total_prev + 1)

medals_per_athlete = Total_prev / (athlete_count + 1)

athlete_change = float(input("Enter athlete change: "))

is_host = int(input("Is host? (0 or 1): "))

# create dataframe
new_data = pd.DataFrame([[
    region, Year, athlete_count, age_median, height_median,
    weight_median, Gold_prev, Silver_prev, Bronze_prev,
    Total_prev, is_zero_prev, rolling_medals, gold_ratio,
    medals_per_athlete, athlete_change, is_host
]], columns=[
    'region', 'Year', 'athlete_count', 'age_median', 'height_median',
    'weight_median', 'Gold_prev', 'Silver_prev', 'Bronze_prev',
    'Total_prev', 'is_zero_prev', 'rolling_medals', 'gold_ratio',
    'medals_per_athlete', 'athlete_change', 'is_host'
])
```

```python
new_data = new_data.reindex(columns=features, fill_value=0)
```

```python
region_freq = joblib.load("region_freq.pkl")
new_data['region'] = new_data['region'].map(region_freq).fillna(0)
```

```python
sfs = joblib.load("sfs_selector.pkl")
new_data_processed = sfs.transform(new_data)
```

```python
prediction = model.predict(new_data_processed)

print("Predicted Output:", prediction)
```
